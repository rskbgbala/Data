```python
#import the required libraries
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
#import the URL and HTML code get to load the variable

url = "https://www.sec.gov/forms"
html_data = requests.get(url).text
```

```python
soup = BeautifulSoup(html_data,"html.parser")
print(soup.prettify())
```

```python
#Got the table from the web site
tables = soup.find_all("table")
len(tables)
```

```python
for i,table in enumerate(tables):
    table_index = i
    print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
#create the index and load data from the list
US_securities_Exchange = pd.DataFrame(columns = ["Number" , "Descripition","Last_upate","SEC_Number","Topics"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    Number = col[0].text
    Descripition = col[1].text
    Last_upate = col[2].text
    SEC_Number= col[3].text
    Topics= col[4].text
    US_securities_Exchange = US_securities_Exchange.append({"Number":Number,"Descripition":Descripition,"Last_upate":Last_upate,"SEC_Number":SEC_Number,"Topics":Topics},ignore_index = True)
```

```python
US_securities_Exchange=pd.DataFrame(US_securities_Exchange)
```

```python
#Remove the next line(\n) from the table
US_securities_Exchange =US_securities_Exchange.replace('\n','',regex=True)
```

```python
US_securities_Exchange
```

```python
#Remove the unwanted text and punctuation from the table
US_securities_Exchange["Number"]=US_securities_Exchange["Number"].replace('Number:','',regex=True)
US_securities_Exchange["Descripition"]=US_securities_Exchange["Descripition"].replace("Description:","",regex=True)
US_securities_Exchange["SEC_Number"]=US_securities_Exchange["SEC_Number"].replace('SEC Number:','',regex=True)
US_securities_Exchange["Last_upate"]=US_securities_Exchange["Last_upate"].replace('Last Updated:','',regex=True)
US_securities_Exchange["Topics"]=US_securities_Exchange["Topics"].replace("Topic\(.*\)\:","",regex=True)
```
